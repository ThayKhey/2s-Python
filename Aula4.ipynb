```python
dados = {
    'Nome':['alice', 'bob', 'charlie', 'david', 'camila'],
    'Idade':[25, 30, 35, 40, 45],
    'Salario':[5000, 6000, 7000, 8000,9000]
}

dados = pd.DataFrame(dados)
dados
```

```python
# janela da média móvel

m = 2

medias_moveis = []

for i in range(len(dados)):
  if i < m - 1:
    medias_moveis.append(None)
  else:
    media = sum(dados['Salario'][i - m + 1:i+1])
    medias_moveis.append(media)

dados['media_movel'] = medias_moveis
```
